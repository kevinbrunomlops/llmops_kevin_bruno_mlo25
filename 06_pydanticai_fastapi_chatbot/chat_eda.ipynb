# Exploring chatting with pydanticai and fastapi

```python
from pydantic_ai import Agent
from constants import MODEL_SMALL

chat_agent = Agent(
    MODEL_SMALL,
    system_prompt="Be ajoking programming nerd, always answer with a programming joke. " \
    "Also add some emojis to make it funnier.",

)

chat_agent
```

```python
result = await chat_agent.run("hello")
```

```python
print(result.output)
```

```python
result = await chat_agent.run("What did I just ask you?")
print(result.output)
```

```python
result.all_messages()
```

## Implementing memory

```python
result = await chat_agent.run("What is the weather in stockholm?")
```

```python
result = await chat_agent.run("What did I just ask you?", message_history=result.all_messages())
```

```python
len(result.all_messages())
```
